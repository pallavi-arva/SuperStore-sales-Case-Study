```python
df_deduplicated['Order Date'] = pd.to_datetime(
    df_deduplicated['Order Date'], 
    format='mixed', 
    errors='coerce'
)
df_deduplicated['Ship Date'] = pd.to_datetime(
    df_deduplicated['Ship Date'], 
    format='mixed',
    errors='coerce'
)
```

```python
df_deduplicated['Order Year'] = df_deduplicated['Order ID'].apply(
    lambda x: int(x.split('-')[1])
)
df_deduplicated['Order Date Year'] = df_deduplicated['Order Date'].dt.year
inconsistencies = df_deduplicated[df_deduplicated['Order Year'] != df_deduplicated['Order Date Year']]
inconsistencies
```

```python
inconsistencies.index
```

```python
df_deduplicated.loc[inconsistencies.index, 'Order Date'] = df_deduplicated.loc[inconsistencies.index].apply(
     lambda row: row['Order Date'].replace(year=row['Order Year']), axis=1
)
impacted_rows = inconsistencies.shape[0]
impacted_order_ids = inconsistencies['Order ID'].nunique()
print(f"Rows impacted by date correction: {impacted_rows}")
print(f"Distinct Order IDs impacted: {impacted_order_ids}")
```

```python
df_deduplicated['Days to Ship'] = (df_deduplicated['Ship Date'] - df_deduplicated['Order Date']).dt.days
df_deduplicated
```

```python
df_deduplicated['Ship Mode'] = df_deduplicated.apply(
    lambda row: 'Same Day' if row['Days to Ship'] == 0 
    else ('Standard Class' if row['Days to Ship'] == 7 
    else row['Ship Mode']), axis=1
)
df_deduplicated
```

```python
df_deduplicated['Quantity'] = pd.to_numeric(df_deduplicated['Quantity'], errors='coerce')
df_deduplicated['Sales Price'] = pd.to_numeric(df_deduplicated['Sales Price'], errors='coerce')
```

```python
# 1) Clean only trailing junk like "\" "/" "." "_" "-" and spaces
df_deduplicated['State'] = (
    df_deduplicated['State']
      .str.strip()
      .str.rstrip(r'\/._- ')   # note: includes backslash
)

# 2) Map abbreviations → full names
state_mapping = {
    'CA': 'California',
    'TX': 'Texas',
    'WA': 'Washington',
    'NY': 'New York',
    'NJ': 'New Jersey'
}
df_deduplicated['State'] = df_deduplicated['State'].replace(state_mapping)
```

```python
df_deduplicated['Original Price'] = df_deduplicated['Sales Price'] / (1 - df_deduplicated['Discount'])
df_deduplicated['Discount Price'] = df_deduplicated['Discount'] * df_deduplicated['Original Price']
df_deduplicated
```

```python
df_deduplicated['Total Sales'] = df_deduplicated['Sales Price'] * df_deduplicated['Quantity']
df_deduplicated['Total Profit'] = df_deduplicated['Profit'] * df_deduplicated['Quantity']
df_deduplicated['Total Discount'] = df_deduplicated['Discount Price'] * df_deduplicated['Quantity']
df_deduplicated
```

```python
df_deduplicated['Shipping Urgency'] = df_deduplicated['Days to Ship'].apply(
    lambda x: 'Immediate' if x == 0 
    else ('Urgent' if 1 <= x <= 3 
    else 'Standard')
)
df_deduplicated
```

```python
df_deduplicated['Days Since Last Order'] = ( 
    df_deduplicated.sort_values(['Customer ID', 'Order Date'])
    .groupby('Customer ID')['Order Date']
    .diff()
    .dt.days
)
df_deduplicated
```

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 3 * IQR
    upper_bound = Q3 + 3 * IQR

    print(f"Detecting outliers in '{column}' using 3 * IQR...")
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"Outliers found: {outliers.shape[0]} rows")
    
    df_cleaned = df[~((df[column] < lower_bound) | (df[column] > upper_bound))]
    
    impacted_rows = outliers.shape[0]
    impacted_order_ids = outliers['Order ID'].nunique()
    print(f"Impacted rows: {impacted_rows}")
    print(f"Distinct impacted Order IDs: {impacted_order_ids}")
    
    return df_cleaned

df_cleaned = remove_outliers(df_deduplicated, 'Sales Price')
df_cleaned = remove_outliers(df_cleaned, 'Profit')
```

```python
df_deduplicated = df_deduplicated.copy() # Shallow Copy

sales_sum  = df_deduplicated.groupby('Customer ID')['Total Sales'].transform('sum')
profit_sum = df_deduplicated.groupby('Customer ID')['Total Profit'].transform('sum')

# use rank to avoid many identical-bin edges; drop duplicates if needed
df_deduplicated.loc[:, 'Customer Sales Quintile'] = (
    pd.qcut(sales_sum.rank(method='first'), 5, labels=False, duplicates='drop') + 1
)
df_deduplicated.loc[:, 'Customer Profit Quintile'] = (
    pd.qcut(profit_sum.rank(method='first'), 5, labels=False, duplicates='drop') + 1
)

cross_grid = pd.crosstab(
    df_deduplicated['Customer Sales Quintile'],
    df_deduplicated['Customer Profit Quintile']
)
print(cross_grid)
```
